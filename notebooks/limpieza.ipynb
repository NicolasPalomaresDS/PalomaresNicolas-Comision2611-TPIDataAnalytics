## Imports
---

```python
import os
from pathlib import Path
import pandas as pd
import numpy as np
from datacleaner import DataCleaner

DATA_DIR = Path("../data/raw")
```

## Data
---

```python
df_ventas = pd.read_csv(DATA_DIR / "ventas.csv")
cleaner_ventas = DataCleaner(df_ventas)
cleaner_ventas.df.set_index("id_venta", inplace=True)

df_clientes = pd.read_csv(DATA_DIR / "clientes.csv")
cleaner_clientes = DataCleaner(df_clientes)
cleaner_clientes.df.set_index("id_cliente", inplace=True)

df_marketing = pd.read_csv(DATA_DIR / "marketing.csv")
cleaner_marketing = DataCleaner(df_marketing)
cleaner_marketing.df.set_index("id_campanha", inplace=True)
```

## 'Ventas'
---

```python
cleaner_ventas.get_summary(name="df_ventas")
```

### Datos nulos

```python
cleaner_ventas.clean_nulls(columns=["precio", "cantidad"], delete=True)
```

### Datos duplicados

```python
cleaner_ventas.clean_dups_by_id()
```

### Corrección de tipos de datos

```python
cleaner_ventas.cast_columns(columns=["precio"], to="float64")
cleaner_ventas.cast_columns(columns=["cantidad"], to="int64")
cleaner_ventas.cast_columns(columns=["fecha_venta"], to="datetime")
```

### Outliers

```python
cleaner_ventas.display_outliers(columns=["precio", "cantidad"], method="iqr")
```

## 'Marketing'
---

```python
cleaner_marketing.get_summary(name="df_marketing")
```

### Corrección de tipos de datos

```python
cleaner_marketing.cast_columns(columns=["fecha_inicio", "fecha_fin"], to="datetime")
```

### Outliers

En este caso no se eliminará ni reemplazará el outlier detectado, ya que es perfectamente normal dado el contexto, y no es un caso extremo.

```python
cleaner_marketing.display_outliers(columns=["costo"], method="iqr")
```

## 'Clientes'
---

```python
cleaner_clientes.get_summary(name="df_clientes")
```

### Corrección de tipos de datos

```python
cleaner_clientes.cast_columns(columns=["ciudad"], to="category")
```

### Outliers

Valores atípicos en `ingresos`: Solamente se tratarán los outliers bajos, ya que teniendo en cuenta que se trata de ciudades en Argentina, esos valores bajos en ingresos no son realistas teniendo en cuenta el salario mínimo.

Los outliers altos se dejarán como están, porque aunque sean casos muy raros o excepcionales, son perfectamente posibles.

```python
cleaner_clientes.display_outliers(columns=["ingresos"], method="iqr")
```

```python
cleaner_clientes.clean_outliers(columns=["ingresos"], method="iqr", side="lower")
```

## Guardar datos limpios en `root/data/clean`
---

Formato `parquet` para mantener los tipos de datos.

```python
CLEAN_DATA_DIR = Path("../data/clean")

cleaner_ventas.df.reset_index(inplace=True)
cleaner_ventas.df.to_parquet(CLEAN_DATA_DIR / "ventas.parquet", index=False)

cleaner_marketing.df.reset_index(inplace=True)
cleaner_marketing.df.to_parquet(CLEAN_DATA_DIR / "marketing.parquet", index=False)

cleaner_clientes.df.reset_index(inplace=True)
cleaner_clientes.df.to_parquet(CLEAN_DATA_DIR / "clientes.parquet", index=False)
```
